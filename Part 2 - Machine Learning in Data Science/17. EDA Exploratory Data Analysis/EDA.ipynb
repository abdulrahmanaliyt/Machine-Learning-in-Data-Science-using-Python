```python
cars = cars.drop(['MSRP','Invoice'], axis=1 )
```

```python
cars = cars.drop_duplicates(keep='first')
```

```python
cars.isnull().sum()
```

```python
cars.dropna(inplace=True)
```

```python
cars_sort = cars.sort_values(by= 'MPG_City', ascending=False)
```

```python
cars_sort.head()
```

```python
cars['Origin'].unique()
```

```python
# Convert continent names into numeric
# Asia into 1, Europe into 2 , USA into 3

def origin_to_num(x):
    if x == 'Asia': return 1
    if x == 'Europe': return 2
    if x == 'USA': return 3
```

```python
# apply the above method on Origin columns
cars['Origin'] = cars['Origin'].apply(origin_to_num)
```
